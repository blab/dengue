```python
import matplotlib.pyplot as plt
import baltic as bt
import pickle
import seaborn as sns
%matplotlib inline

import matplotlib.patches as mpatches
```

```python
annotated_tree = '../data/trees/genotype_tree.nexus'
tree = bt.loadNexus(annotated_tree, absoluteTime=False)
```

```python
colors = pickle.load(open('./colors.p', 'rb'))
```

```python
efficacy = {'DENV1': .547,
            'DENV1_I': .556,
            'DENV1_IV': .544,
            'DENV2': .346,
            'DENV2_ASIANI': .212,
            'DENV2_COSMOPOLITAN': .427,
            'DENV3': .651,
            'DENV3_II': 0.876,
            'DENV3_I': 0.612,
            'DENV3_III': 0.612,
            'DENV4_I': .587,
            'DENV4_II': .839,
           'DENV4': .724}

for k in tree.leaves:
    if 'genotype' in k.traits and k.traits['genotype'] in efficacy:
        k.traits['efficacy'] = efficacy[k.traits['genotype']]
    else:
        k.traits['efficacy'] = efficacy[k.traits['serotype']]
```

```python
def plot_tree(tree, labels=False, colorby='genotype', colors=colors):
    branchWidth=2 ## default branch width
    ll = tree

    plt.yticks(size=0)
    
    for k in ll.Objects: ## iterate over objects in tree
        x=k.x ## or use absolute time instead
        y=k.y ## get y position from .drawTree that was run earlier, but could be anything else

        xp=k.parent.x ## get x position of current object's parent
        if x==None: ## matplotlib won't plot Nones, like root
            x=0.0
        if xp==None:
            xp=x
            
        if k.traits.has_key(colorby):
            try:
                c = colors[k.traits[colorby]]
            except:
                c = colors(k.traits[colorby])
            z = 10
        else:
            c = 'gray'
            z = 5
            
        if isinstance(k,bt.leaf) or k.branchType=='leaf': ## if leaf...
            s=50 ## tip size can be fixed
            s=50-30*k.height/ll.treeHeight
            
            if 'colorby' in k.traits:
                label = k.traits[colorby]#k.traits[colorby].replace('_', ' ')
            else:
                label = ''
            plt.scatter(x,y,s=s,facecolor=c,edgecolor='none',zorder=z+1, label=label) ## plot circle for every tip
            plt.scatter(x,y,s=s+0.8*s,facecolor='k',edgecolor='none',zorder=z) ## plot black circle underneath
                        
        elif isinstance(k,bt.node) or k.branchType=='node': ## if node...
            plt.plot([x,x],[k.children[-1].y,k.children[0].y],lw=branchWidth,color=c,ls='-',zorder=z-1)
            
        plt.plot([xp,x],[y,y],lw=branchWidth,color=c,ls='-',zorder=z-1)
    plt.xlabel('Substitutions per site')
    
    ax.spines['top'].set_visible(False) ## no axes
    ax.spines['right'].set_visible(False)
    ax.spines['left'].set_visible(False)
    ax.xaxis.grid(True)
#     ax.spines['bottom'].set_visible(False)

    x0,x1 = plt.xlim()
    plt.xlim((x0, x1*1.1))
```

```python
sns.set(style='white', font_scale=1.2)
fig, (ax, cax) = plt.subplots(ncols=2,nrows=1,figsize=(7.5, 7.5), gridspec_kw={'width_ratios': (1, 0.03)})
plt.sca(ax)
plot_tree(tree,labels=True,colorby='efficacy',colors=plt.cm.YlGn)
# ax.legend(loc=(0.9,0.4), handles=[labels], ncol=1)
sm = plt.cm.ScalarMappable(cmap=plt.cm.YlGn, norm=plt.Normalize(vmin=0, vmax=1))
sm._A = []
plt.colorbar(mappable=sm,label='Vaccine Efficacy',cax=cax )
plt.tight_layout()
plt.savefig('./png/genotype_ve_tree.png', dpi=300, bbox_inches='tight')
```
